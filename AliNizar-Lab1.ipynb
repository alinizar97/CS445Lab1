```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Wine Quality dataset
# Make sure the 'winequality-red.csv' file is in your working directory
df = pd.read_csv("winequality-red.csv", sep=';')

# Display the first few rows to understand the structure of the data
print(df.head())
```

```python
# Group data by wine quality and calculate the average alcohol percentage for each group
avg_alcohol = df.groupby('quality')['alcohol'].mean()

# Create a bar plot for the average alcohol content by wine quality
plt.figure(figsize=(10, 6))
avg_alcohol.plot(kind='bar', color='skyblue')
plt.title('Average Alcohol Content by Wine Quality', fontsize=14)
plt.xlabel('Wine Quality', fontsize=12)
plt.ylabel('Average Alcohol (%)', fontsize=12)
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()

# Save the bar chart as an image
plt.savefig("AliNizar_AlcoholBarChart.png")
plt.show()
```

```python
# Sort the DataFrame by alcohol content to show a proper line trend
sorted_df = df.sort_values('alcohol')

# Create a line graph for pH levels across increasing alcohol content
plt.figure(figsize=(10, 6))
plt.plot(sorted_df['alcohol'], sorted_df['pH'], color='green')
plt.title('pH Level vs. Alcohol Content', fontsize=14)
plt.xlabel('Alcohol (%)', fontsize=12)
plt.ylabel('pH Level', fontsize=12)
plt.tight_layout()

# Save the line chart as an image
plt.savefig("AliNizar_PHvsAlcohol_LineChart.png")
plt.show()
```

```python
# Count how many times each wine quality rating appears in the dataset
quality_counts = df['quality'].value_counts()

# Create a pie chart to show the distribution of wine quality ratings
plt.figure(figsize=(8, 8))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Wine Quality Ratings', fontsize=14)

# Save the pie chart as an image
plt.savefig("AliNizar_QualityDistribution_PieChart.png")
plt.show()
```

```python
# Create a box plot to visualize the distribution of residual sugar in the wines
plt.figure(figsize=(8, 6))
plt.boxplot(df['residual sugar'])
plt.title('Box Plot of Residual Sugar', fontsize=14)
plt.ylabel('Residual Sugar (g/dm³)', fontsize=12)
plt.tight_layout()

# Save the box plot as an image
plt.savefig("AliNizar_ResidualSugar_BoxPlot.png")
plt.show()
```
